```python
storage_account_name = 'provide the storage account name'
client_id = 'the application id registered in azure ad directory'
tenant_id = 'the directory id for the registered application'
client_secret = 'the secret for the application'
```

```python
dbutils.secrets.get(scope='adls2-access-scope',key='adls2-secret')
```

```python
configs = {
    'fs.azure.account.auth.type': 'OAuth',
    'fs.azure.account.oauth.provider.type': 'org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider',
    'fs.azure.account.oauth2.client.id': f"{client_id}",
    'fs.azure.account.oauth2.client.secret': f'{client_secret}',
    'fs.azure.account.oauth2.client.endpoint': f"https://login.microsoftonline.com/{tenant_id}/oauth2/token"
}
```

```python
container_name = 'cmpart'
dbutils.fs.mount(
    source = f"abfss://{container_name}@{storage_account_name}.dfs.core.windows.net",
    mount_point = f"/mnt/{container_name}",
    extra_configs = configs
)
```

```python
dbutils.fs.ls('/mnt/cmpart')
```
